```python
#一切类必需要初始化
#在类中所有函数的第一参数都是标识量不是参数
#若一个函数要多次使用，那么将其统一放在初始化函数中
class Joker:
    def __init__(self,num1,num2):
        print('我开始初始化')
        self.num1=num1
        self.num2=num2
    def print_(self,name):
        print('hello word',name)
    def sum(self):
        return self.num1 + self.num2
    def cheng(self):
        return self.num1 * self.num2
```

```python
class account:
    def __init__(self,id_ = 0,balance = 100,n_rate = 0):
        self.__id_ = id_
        self.__balance = balance
        self.__n_rate = n_rate
        print('账户的id是：',self.__id_)
    def yuelilv(self):
        self.yll=self.__balance*(self.__n_rate/12)/100
        print('月利率为：',self.yll)
    def yuelixi(self):
        self.ylx=self.__balance*self.yll
        print('月利息为：',self.ylx)
    def withdraw(self):
        num=eval(input('请输入取钱金额：'))
        self.__balance=self.__balance-num
        print('账号里还有%d美元'%self.__balance)
    def deposit(self):
        num1=eval(input('请输入存钱金额：'))
        self.__balance=self.__balance+num1
        print("账户剩余%d美元"%self.__balance)
```

```python
acc=account(1122,20000,4.5)
```

```python
acc.yuelilv()
acc.yuelixi()
acc.withdraw()
acc.deposit()
```

```python
import math
class nbian:
    def __init__(self):
        pass
    def n_bian(self,n=3,side=1,x=0,y=0):
        self.__n=n
        self.__side=side
        self.__x=x
        self.__y=y
    def zhouchang(self):
        zc=self.__n*self.__side
        print('周长为：',zc)
    def  area(self):
        num=math.tan(math.pi/self.__n)
        mj=(self.__n*self.__side**2)/4*num
        print('面积为：',mj)
```

```python
nn=nbian()
```

```python
nn.n_bian(8,4)
nn.zhouchang()
nn.area()
```

```python
class LinearEquation:
    def __init__(self,a,b,c,d,e,f):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
        self.e=e
        self.f=f
    def isSolvable(self):
        if self.a*self.d - self.b*self.c != 0:
            x=(self.e*self.d-self.b*self.f)/(self.a*self.d-self.b*self.c)
            y=(self.a*self.f-self.e*self.c)/(self.a*self.d-self.b*self.c)
            print(x,y,'True')
        else:
            print('无解')
```
